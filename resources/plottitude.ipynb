```python
import pandas as pd
```

```python
df= pd.read_csv("cities.csv")
df
```

```python
df.to_html('data.html')
```

```python
df_humid = df.loc[(df['Humidity'] > 90) | (df['Humidity'] < 20)]
df_humid
```

df_lowhumid = df.loc[df['Humidity'] < 20]
df_lowhumid

```python
df_clouds = df.loc[(df['Cloudiness'] == 0) | (df['Cloudiness'] == 100)]
df_clouds
```

df_maxclouds = df.loc[df['Cloudiness'] == 100]
df_maxclouds

```python
df_temp = df.loc[(df['Max Temp'] > 99) | (df['Max Temp'] < 40)]
df_temp
```

```python
df_wind = df.loc[(df['Wind Speed'] > 20) | (df['Wind Speed'] < 5)]
df_wind
```

```python
df_humid.to_html('humid.html', border=0, classes='table-striped table')
df_clouds.to_html('clouds.html',border=0, classes='table-striped table')
df_temp.to_html('temp.html',border=0, classes='table-striped table')
df_wind.to_html('wind.html',border=0, classes='table-striped table')
```
